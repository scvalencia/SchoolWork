```python
def gaussian(x,x0,sigma):
    return np.exp(-np.power((x - x0)/(2.*sigma), 2.))
```

```python
xx, yy=np.mgrid[-3:5:0.1, -2:4:0.1]
```

```python
print yy
```

```python
z1=gaussian(xx,1.0,2.0)
z2=gaussian(yy,1.0,2.0)
```
