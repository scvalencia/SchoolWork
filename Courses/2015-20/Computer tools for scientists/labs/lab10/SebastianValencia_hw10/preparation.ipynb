```python
x = linspace(0, 2*pi, 10)
```

```python
x
```

```python
y = sin(x)
```

```python
y
```

```python
plot(x, y)
```

```python
plot(x, y, c = "green")
```

```python
plot(x, y, c = "red", label = "sin(x)")
legend(loc=0)
```

```python
z = cos(x)
```

```python
z
```

```python
plot(x, z, c = "green", label = "cos(x)")
plot(x, y, c = "red", label = "sin(x)")
xlabel("x")
ylabel("f(x)")
legend(loc=0)
```

```python
plot(x, z, c = "green", label = "$\cos(x)$")
plot(x, y, c = "red", label = "$\sin(x)$")
xlabel("$x$")
ylabel("$f(x)$")
legend(loc=0)
```

```python
scatter(x, z, c = "green", label = "cos(x)")
xlabel("x")
ylabel("f(x)")
legend(loc=0)
```

```python
scatter(x, y, c = "green", label = "$\sin(x)$", s = 1)
xlabel("x")
ylabel("f(x)")
legend(loc=0)
```
